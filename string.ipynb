```python
titanic=pd.read_csv("titanic (1).csv")
titanic.head()
#only name is string in titanic dataset then i can only operate in using name
```

```python
titanic["Name"].str.lower()
```

```python
titanic["Name"].str.capitalize()
```

```python
titanic["Name"].str.title()
```

almost all python string function is working here

```python
#find the name length of passenger and find the maximium length of name
titanic["Name"].str.len().sort_values(ascending=False).max()
```

```python
#find surname of each passenger
#split()
titanic["Surname"]=titanic["Name"].str.split(",").str.get(0)
titanic["Surname"]
```

```python
titanic[["title","firstName"]]=titanic["Name"].str.split(",").str.get(1).str.strip().str.split(" ",n=1,expand=True)
titanic["firstName"]
```

```python
titanic["title"].value_counts()
```

```python
titanic["title"].str.replace("Miss","Mrs")
```

**Filtering**

```python
titanic[titanic["firstName"].str.startswith("A")]
```

```python
#isdigit and isalpha
titanic[titanic["Name"].str.isalpha()]
```

appling regix

```python
# applying regex
# contains
# search john -> both case
titanic[titanic["Name"].str.contains("john",case=False)].head()
```

```python
#find lastname with start and end with vowel
titanic[titanic["Surname"].str.contains("^[AEIOUaeiou].+[AEIOUaeiou]$")]
```
