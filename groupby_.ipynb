```python
#groupby functions in pandas (it's apply only categorical data)
genres=movies.groupby("Genre")        #group by genres
```

```python
genres.sum()    #sum of all numerical column
```

```python
genres.min()    #minimium value in each numerical column
```

```python
#  Q. find the top 3 geners by votes
genres["No_of_Votes"].sum().sort_values(ascending=False).head(3)
```

```python
 # Q. find the genres with highest imdb rating
 genres["IMDB_Rating"].max().head(1)
```

```python
# Q. find director with most populor by voting
dir=movies.groupby("Director")
dir["No_of_Votes"].sum().sort_values(ascending=False).head(1).index[0]
```

```python
#Q. find the highest rating movies in each genres
newdf=genres["Series_Title","IMDB_Rating"]
newdf.max()
```

```python
#agg method (Aggrigation)
#columns name and it's operations of each group
genres.agg({"Runtime":"sum",
            "IMDB_Rating":"mean",
            "No_of_Votes":"sum",
            "Gross":"mean"})
```

```python
#now mean max and sum in each numerical column
genres.agg(["mean","max","sum"])
```

```python
#merge both
genres.agg({"Runtime":["mean","max","sum"],
           "No_of_Votes":["mean","max","min"],
           "Gross":"sum"})
```
