```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

#**pandas plot()**

```python
# on a series
s=pd.Series([1,2,3,4,5,6,7,8])
colors=["red","orange","blue","yellow","black","green","maroon"]
s.plot(kind="pie",colors=colors)
```

**plot using a dataset**

```python
import seaborn as sns
tip = sns.load_dataset("tips")
tip.head()
```

```python
tip["size"]=tip["size"]*100
tip.head()
```

```python
#now plot a scatter on this dataframe
tip.plot(kind="scatter",x="total_bill",y="tip",marker="+",s=tip["size"],c="sex",figsize=(8,4),cmap="viridis")
# s=size of marker
# c= filter according to given different value
# cmap means fill different color into different filter marker
```

```python
# 2D plot using a stock dataset

stocks = pd.read_csv('https://raw.githubusercontent.com/m-mehdi/pandas_tutorials/main/weekly_stocks.csv')
stocks.head()
```

```python
stocks.shape
```

```python
#this is only for prectice
import datetime as dt
x=pd.to_datetime(stocks["Date"]).dt.month_name()
stocks["Date"].groupby(x)
stocks
```

```python
stocks.plot(kind="line",x="Date",y=["MSFT","FB","AAPL"],figsize=(8,6))
#figsize means size of figure
```

```python
# another way to stock analysis
stocks["FB"].plot(kind="line",x=stocks["Date"])
```

```python
stocks[["Date","FB","MSFT","AAPL"]].plot(kind="line",x="Date")
```

```python
stocks[["Date","FB","MSFT","AAPL"]].plot(kind="bar",x="Date",figsize=(12,10),)
```

** bar chat single multiple and horizontal virtical**

```python
#it is somthing else out of understanding
tip.groupby("sex").plot(kind="bar",x="total_bill",y="tip")
```

```python
#ploting graph for average bill given by male and female
tip.groupby("sex")["total_bill"].mean().plot(kind="bar",color=["blue","green"])
```
