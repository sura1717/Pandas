```python
#inner join
#here students is left table, nov is right table, how means which type of join, on means on the basis of(a common column)
students.merge(nov,how="inner",on="student_id")
```

```python
#left join---> all value which is in the common and left table but not in right table
courses.merge(regs,how="left",on="course_id")
```

```python
#right join---> which is exist in right table but not in left
students.merge(regs,how="right",on="student_id")
```

```python
#full outer join--->common or own presonal data both are seen
students.merge(regs,how="outer",on="student_id")
```

***prectice question***

```python
# 1. find total revenue generated
total=regs.merge(courses,how="inner",on="course_id")["price"].sum()
total
```

```python
# 2.  find month by month revenue
tempdf=pd.concat([nov,dec],keys=["nov","dec"]).reset_index()
ans=tempdf.merge(courses,on="course_id").groupby("level_0").sum()
ans
```

```python
# 3.  print the registration table--> cols--name,course,price
ans=regs.merge(students,on="student_id").merge(courses,on="course_id")
ans[["name","course_name","price"]]
```

```python
# 4.  find student who enroll both the month
stid=np.intersect1d(nov["student_id"],dec["student_id"])
students[students["student_id"].isin(stid)]
```

```python
# 5.  find courses that get no enrollment
tempid=np.setdiff1d(courses["course_id"],regs["course_id"])
courses[courses["course_id"].isin(tempid)]
```

```python
# 6.  print student name ,partinar name for all enrolled student
both=students.merge(students,how="inner",left_on="partner",right_on="student_id")
both[["name_x","name_y"]]
```
